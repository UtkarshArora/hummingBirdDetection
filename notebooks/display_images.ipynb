```python
import os
from PIL import Image
import matplotlib.pyplot as plt
print("libraries imported successfully")
```

```python
image_dir = "../data/raw/Dataset_HBird/train/images"
```

```python
if not os.path.exists(image_dir):
    print(f"The path '{image_dir}' does not exist")
else:
    print("the path exists")
```

```python
all_files = os.listdir(image_dir)
```

```python
all_files
```

```python
image_extensions = (".img", ".jpg", ".jpeg", ".tiff", ".bmp")
```

```python
image_extensions
```

```python
image_files = [f for f in all_files if f.lower().endswith(image_extensions)]
```

```python
image_files
```

```python
if not image_files:
    print(f"No images found in '{image_dir}' with the image_extensions")
else:
    print(f"Found {len(image_files)} in {image_dir}, here are the first 5 {image_files[:5]}")
```

```python
plt.figure(figsize = (15, 8))
```

```python
num_images_to_display = 5
for i, img_file in enumerate(image_files[:num_images_to_display]):
    full_image_path = os.path.join(image_dir, img_file)
    try:
        img = Image.open(full_image_path)
        plt.subplot(5, 1, i+1)
        plt.imshow(img)
        plt.title(img_file, fontsize=8)
        plt.axis('off')
    except Exception as e:
        print(f"Could not display image file: '{img_file}' due to {e}")
plt.tight_layout()
plt.show()
```
